```python
datFile =  "shape_predictor_68_face_landmarks.dat"
```

```python
def downsampling(img):
    g_down = []
    g_down.append(img)

    for i in range(3):
        tmp = cv2.pyrDown(img)
        g_down.append(tmp)
        img = tmp

    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return g_down[1]
```

```python
def show_raw_detection(image, detector, predictor):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    rects = detector(image, 1)

    for (i, rect) in enumerate(rects):
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        (x,y,w,h) = face_utils.rect_to_bb(rect)
        cv2.rectangle(image, (x,y), (x+w,y+h),(0,255,0),2)

        cv2.putText(image, "Face #{}".format(i+1), (x-10,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)

        for (x,y) in shape:
            cv2.circle(image, (x,y), 1, (0,0,255), -1)

    cv2.imshow("",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
from PIL import Image

def read_images_from_folder(folder_path):
    image_list = []

    for filename in os.listdir(folder_path):
        # 이미지 파일인지 확인
        if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
            image_path = os.path.join(folder_path, filename)
            try:
                # 이미지 파일 열기
                img = Image.open(image_path)
                # 이미지 리스트에 추가
                image_list.append(img)
            except Exception as e:
                # 이미지 파일 열기에 실패한 경우 예외 처리
                print(f"Failed to open image: {image_path}\nError: {str(e)}")

    return image_list
```

```python
# 이미지가 있는 폴더 경로 설정
folder_path = "sad/"

# 폴더 내 이미지 읽어오기
images = read_images_from_folder(folder_path)
# print(images)
```

```python
import uuid
import warnings
```

```python
def generate_random_filename(extension):
    random_name = uuid.uuid4().hex
    filename = f"{random_name}.{extension}"
    return filename
```

```python
for image in images:
    image_np = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    downsample_face = downsampling(image_np)

    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(datFile)
    image = imutils.resize(downsample_face, width=500)
#     show_raw_detection(image, detector, predictor)

    landmarks = extract_face_landmarks(image)
    
    if not detector(image, 1):
        warnings.warn("No face detected.")
        None
    else:
        fig = plt.figure(figsize=(15, 5))
        ax = fig.add_subplot(1, 3, 2)
        ax.scatter(landmarks[:, 0], -landmarks[:, 1], alpha=0.8)
    
        # 결과 이미지 파일 경로
        output_folder = "sad_new/"

        output_file = os.path.join(output_folder,generate_random_filename("jpg"))

        fig.savefig(output_file)
        plt.close(fig)
```

```python
def crop_images_in_folder(input_folder, output_folder, x, y,w,h):
    # 출력 폴더가 없는 경우 생성
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 입력 폴더의 이미지 파일들을 처리
    for filename in os.listdir(input_folder):
        # 이미지 파일인지 확인
        if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
            image_path = os.path.join(input_folder, filename)

            # 이미지 불러오기
            image = cv2.imread(image_path)

            # 이미지가 정상적으로 불러와졌을 경우에만 처리
            if image is not None:
                # 이미지 자르기
                cropped_image = image[y:y+h, x:x+w]

                # 저장할 파일 경로 생성
                output_path = os.path.join(output_folder, filename)

                # 이미지 저장
                cv2.imwrite(output_path, cropped_image)

#                 print(f"Cropped image saved: {output_path}")

# 입력 폴더와 출력 폴더 경로 설정
input_folder = "sad_new/"
output_folder = "sad_final/"

# 이미지 자르기 및 저장
x,y = 590, 50
w,h = 370, 400
crop_images_in_folder(input_folder, output_folder, x, y,w,h)
```
